# MERGING CSV FILES

```python
import os
import pandas as pd

# Path to the folder containing CSV files
folder_path = r"C:\Users\DELL\Downloads\INDIA DATASET\temps"

# Function to merge CSV files in a folder
def merge_csv_files(folder_path, output_file):
    # Get list of CSV files in the folder
    csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
    
    # Initialize an empty DataFrame to store merged data
    merged_data = pd.DataFrame()
    
    # Iterate over each CSV file and merge its data into the DataFrame
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        data = pd.read_csv(file_path)
        merged_data = pd.concat([merged_data, data])
    
    # Add a new column "site name" with value "Golden temple"
    #merged_data['site name'] = 'bangalore 2'
    
    # Save merged data to a new CSV file
    merged_data.to_csv(output_file, index=False)

# Output file path
output_file = r"C:\Users\DELL\Downloads\INDIA DATASET\temps\merged_datas.csv"

# Call the function to merge CSV files
merge_csv_files(folder_path, output_file)

print("CSV files merged successfully!")
```

```python
import pandas as pd


# Convert 'End date' column to datetime dtype
df['End date'] = pd.to_datetime(df['End date'])

# Define your function to get season
def get_season(date):
    if (date.month == 3 and date.day >= 20) or (date.month == 6 and date.day < 21):
        return 'Spring'
    elif (date.month == 6 and date.day >= 21) or (date.month == 9 and date.day < 22):
        return 'Summer'
    elif (date.month == 9 and date.day >= 22) or (date.month == 12 and date.day < 21):
        return 'Autumn'
    else:
        return 'Winter'

# Apply the function to create a new column representing seasons
df['Season'] = df['End date'].apply(get_season)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
# Assuming your dataset is named 'data.csv' and contains a column named 'DateTime' for date/time and 'PM2.5 (µg/m³)' for PM2.5 data
#data = pd.read_csv('data.csv')

# Convert 'DateTime' column to datetime type
aqi22['Timestamp'] = pd.to_datetime(aqi22['Timestamp'])

# Set 'DateTime' column as index
aqi22.set_index('Timestamp', inplace=True)

# Resample the data to monthly frequency and calculate the mean
monthly_avg = aqi22['PM2.5 (µg/m³)'].resample('D').mean()

# Plotting the monthly average PM2.5 data
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg.values, linestyle='-', marker='o', color='b')
plt.title('Daily Average PM2.5 Levels')
plt.xlabel('Date')
plt.ylabel('PM2.5 (µg/m³)')
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load your dataset
# Assuming your dataset is named 'data.csv' and contains a column named 'DateTime' for date/time and 'PM2.5 (µg/m³)' for PM2.5 data
#data = pd.read_csv('data.csv')

# Convert 'DateTime' column to datetime type
#data['DateTime'] = pd.to_datetime(data['DateTime'])

# Set 'DateTime' column as index
#data.set_index('DateTime', inplace=True)

# Calculate daily average PM2.5 data
daily_avg = aqi22['PM2.5 (µg/m³)'].resample('D').mean().dropna()

# Plot ACF
plt.figure(figsize=(12, 6))
plot_acf(daily_avg, lags=30, alpha=0.05)
plt.title('Autocorrelation Function (ACF) for Daily Average PM2.5')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

# Plot PACF
plt.figure(figsize=(12, 6))
plot_pacf(daily_avg, lags=30, alpha=0.05)
plt.title('Partial Autocorrelation Function (PACF) for Daily Average PM2.5')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load your dataset
# Assuming your dataset is named 'data.csv' and contains a column named 'DateTime' for date/time and 'PM2.5 (µg/m³)' for PM2.5 data
#data = pd.read_csv('data.csv')

# Convert 'DateTime' column to datetime type
#data['DateTime'] = pd.to_datetime(data['DateTime'])

# Set 'DateTime' column as index
#data.set_index('DateTime', inplace=True)

# Calculate daily average PM2.5 data
daily_avg = aqi22['PM2.5 (µg/m³)'].resample('D').mean().dropna()

# Perform decomposition
decomposition = seasonal_decompose(daily_avg, model='additive')

# Plot the decomposition
plt.figure(figsize=(12, 10))

plt.subplot(411)
plt.plot(daily_avg, label='Original')
plt.legend(loc='upper left')
plt.title('Original Series')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.title('Trend Component')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.title('Seasonal Component')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```
