```python
import pandas as pd
from sklearn.impute import KNNImputer



# Select columns with missing values for imputation
cols_to_impute = ['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO (µg/m³)', 'NO2 (µg/m³)',
       'NOx (ppb)', 'NH3 (µg/m³)', 'SO2 (µg/m³)', 'CO (mg/m³)',
       'Ozone (µg/m³)', 'Benzene (µg/m³)', 'Toluene (µg/m³)', 'AT (°C)',
       'RH (%)', 'WS (m/s)', 'WD (deg)', 'TOT-RF (mm)', 'SR (W/mt2)',
       'BP (mmHg)']

# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=5)  # You can adjust the number of neighbors as needed

# Fit and transform the imputer on selected columns
imputed_values = imputer.fit_transform(data[cols_to_impute])

# Create a dataframe from the imputed values
imputed_df = pd.DataFrame(imputed_values, columns=cols_to_impute, index=data.index)

# Replace the missing values in the original dataframe with the imputed values
data[cols_to_impute] = imputed_df

# Now you have the original dataframe with missing values imputed
print(data.head())
```

```python
data_interpolated = data.copy()
```

```python
# Calculate quartiles and IQR
Q1 = data_interpolated['CO (mg/m³)'].quantile(0.25)
Q3 = data_interpolated['CO (mg/m³)'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outlier detection
lower_bound = Q1 - 3 * IQR
upper_bound = Q3 + 3 * IQR

# Identify outliers
outliers = data_interpolated['CO (mg/m³)'][(data_interpolated['CO (mg/m³)']< lower_bound) | data_interpolated['CO (mg/m³)'] > upper_bound]

print("Outliers detected using IQR method:")
print(outliers)
```

```python
df = data_interpolated.copy()
```

```python
df["PM10_24hr_avg"] = df.groupby("name")["PM10 (µg/m³)"].rolling(window = 24, min_periods = 16).mean().values
df["PM2.5_24hr_avg"] = df.groupby("name")["PM2.5 (µg/m³)"].rolling(window = 24, min_periods = 16).mean().values
df["SO2_24hr_avg"] = df.groupby("name")["SO2 (µg/m³)"].rolling(window = 24, min_periods = 16).mean().values
df["NO2_24hr_avg"] = df.groupby("name")["NO2 (µg/m³)"].rolling(window = 24, min_periods = 16).mean().values
df["CO_8hr_max"] = df.groupby("name")["CO (mg/m³)"].rolling(window = 8, min_periods = 1).max().values
df["O3_8hr_max"] = df.groupby("name")["Ozone (µg/m³)"].rolling(window = 8, min_periods = 1).max().values
df["NH3_24hr_avg"] = df.groupby("name")['NH3 (µg/m³)'].rolling(window = 24, min_periods = 16).mean().values
```

```python
aqi22 = df.copy()
```

```python
## PM2.5 Sub-Index calculation
def get_PM25_subindex(x):
    if x <= 30:
        return x * 50 / 30
    elif x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0

aqi22["PM2.5_SubIndex"] = aqi22['PM2.5_24hr_avg'].apply(lambda x: get_PM25_subindex(x))
```

```python
## PM10 Sub-Index calculation
def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x <= 100:
        return x
    elif x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x <= 350:
        return 200 + (x - 250)
    elif x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0

aqi22["PM10_SubIndex"] = aqi22['PM10_24hr_avg'].apply(lambda x: get_PM10_subindex(x))
```

```python
## SO2 Sub-Index calculation
def get_SO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 380:
        return 100 + (x - 80) * 100 / 300
    elif x <= 800:
        return 200 + (x - 380) * 100 / 420
    elif x <= 1600:
        return 300 + (x - 800) * 100 / 800
    elif x > 1600:
        return 400 + (x - 1600) * 100 / 800
    else:
        return 0

aqi22["SO2_SubIndex"] = aqi22['SO2_24hr_avg'].apply(lambda x: get_SO2_subindex(x))
```

```python
## NOx Sub-Index calculation
def get_NO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 180:
        return 100 + (x - 80) * 100 / 100
    elif x <= 280:
        return 200 + (x - 180) * 100 / 100
    elif x <= 400:
        return 300 + (x - 280) * 100 / 120
    elif x > 400:
        return 400 + (x - 400) * 100 / 120
    else:
        return 0

aqi22["NO2_SubIndex"] = aqi22['NO2_24hr_avg'].apply(lambda x: get_NO2_subindex(x))
```

```python
## CO Sub-Index calculation
def get_CO_subindex(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0

aqi22["CO_SubIndex"] = aqi22['CO_8hr_max'].apply(lambda x: get_CO_subindex(x))
```

```python
## O3 Sub-Index calculation
def get_O3_subindex(x):
    if x <= 50:
        return x * 50 / 50
    elif x <= 100:
        return 50 + (x - 50) * 50 / 50
    elif x <= 168:
        return 100 + (x - 100) * 100 / 68
    elif x <= 208:
        return 200 + (x - 168) * 100 / 40
    elif x <= 748:
        return 300 + (x - 208) * 100 / 539
    elif x > 748:
        return 400 + (x - 400) * 100 / 539
    else:
        return 0

aqi22["O3_SubIndex"] = aqi22['O3_8hr_max'].apply(lambda x: get_O3_subindex(x))
```

```python
## NH3 Sub-Index calculation
def get_NH3_subindex(x):
    if x <= 200:
        return x * 50 / 200
    elif x <= 400:
        return 50 + (x - 200) * 50 / 200
    elif x <= 800:
        return 100 + (x - 400) * 100 / 400
    elif x <= 1200:
        return 200 + (x - 800) * 100 / 400
    elif x <= 1800:
        return 300 + (x - 1200) * 100 / 600
    elif x > 1800:
        return 400 + (x - 1800) * 100 / 600
    else:
        return 0

aqi22["NH3_SubIndex"] = aqi22["NH3_24hr_avg"].apply(lambda x: get_NH3_subindex(x))
```

```python
## AQI bucketing
def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN

aqi22["Checks"] = (aqi22["PM2.5_SubIndex"] > 0).astype(int) + \
                (aqi22["PM10_SubIndex"] > 0).astype(int) + \
                (aqi22["SO2_SubIndex"] > 0).astype(int) + \
                (aqi22["NO2_SubIndex"] > 0).astype(int) + \
                (aqi22["CO_SubIndex"] > 0).astype(int) + \
                (aqi22["NH3_SubIndex"] > 0).astype(int) + \
                (aqi22["O3_SubIndex"] > 0).astype(int)

aqi22["AQI_calculated"] = round(aqi22[["PM2.5_SubIndex", "PM10_SubIndex", "SO2_SubIndex", "NO2_SubIndex",
                                "CO_SubIndex", "O3_SubIndex","NH3_SubIndex"]].max(axis = 1))
aqi22.loc[aqi22["PM2.5_SubIndex"] + aqi22["PM10_SubIndex"] <= 0, "AQI_calculated"] = np.NaN
aqi22.loc[aqi22.Checks < 3, "AQI_calculated"] = np.NaN

aqi22["AQI_bucket_calculated"] = aqi22["AQI_calculated"].apply(lambda x: get_AQI_bucket(x))
aqi22[~aqi22.AQI_calculated.isna()].head(13)
```
