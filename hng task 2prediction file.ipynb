## TRAIN MODEL

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import RandomOverSampler

# Step 2: Preprocess Data
newdf.index = pd.to_datetime(newdf.index)

# Step 3: Prepare the time series data
# Create time-based features
newdf['year'] = newdf.index.year
newdf['month'] = newdf.index.month
newdf['day'] = newdf.index.day
newdf['dayofweek'] = newdf.index.dayofweek

# Define features and target
features = ['tempmax', 'tempmin', 'temp', 'humidity', 'precip', 'precipprob', 
            'precipcover', 'windgust', 'windspeed', 'winddir', 'sealevelpressure', 
            'cloudcover', 'visibility', 'year', 'month', 'day', 'dayofweek']
target = 'flood'

X = newdf[features]
y = newdf[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Handle Imbalance using RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_scaled, y)

# Step 5: Use Stratified K-Fold Cross-Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in skf.split(X_resampled, y_resampled):
    X_train, X_test = X_resampled[train_index], X_resampled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    # Step 6: Build and Train the Logistic Regression Model
    logreg_model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
    logreg_model.fit(X_train, y_train)
    logreg_y_pred = logreg_model.predict(X_test)

    # Step 7: Evaluate the Model
    print("Logistic Regression Model")
    print(confusion_matrix(y_test, logreg_y_pred))
    print(classification_report(y_test, logreg_y_pred))

# Step 8: Save the trained model for future predictions
import joblib
joblib.dump(logreg_model, 'logistic_regression_flood_model.pkl')
```

## PREDICT PROBAB;E DATE

```python
# Import necessary libraries
import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler

# Load the future weather data
future_weather_data = pd.read_csv('Downloads\lagos nigeria 2024-07-07 to 2024-12-31.csv')

# Ensure the datetime column is correctly formatted with dayfirst=True
future_weather_data['datetime'] = pd.to_datetime(future_weather_data['datetime'], dayfirst=True)
future_weather_data.set_index('datetime', inplace=True)

# Create time-based features for future data
future_weather_data['year'] = future_weather_data.index.year
future_weather_data['month'] = future_weather_data.index.month
future_weather_data['day'] = future_weather_data.index.day
future_weather_data['dayofweek'] = future_weather_data.index.dayofweek

# Define features
features = ['tempmax', 'tempmin', 'temp', 'humidity', 'precip', 'precipprob', 
            'precipcover', 'windgust', 'windspeed', 'winddir', 'sealevelpressure', 
            'cloudcover', 'visibility', 'year', 'month', 'day', 'dayofweek']

# Ensure future data aligns with the features used in the model
X_future = future_weather_data[features]

# Load the trained model and scaler
logreg_model = joblib.load('logistic_regression_flood_model.pkl')
scaler = joblib.load('scaler.pkl')

# Scale the future data
X_future_scaled = scaler.transform(X_future)

# Predict flood incidences
future_flood_prediction = logreg_model.predict(X_future_scaled)

# Find the next probable date of flood
next_flood_date = future_weather_data[future_flood_prediction == 1].index.min()
print("Next probable date of flood:", next_flood_date)
```
