```python
import re

def convert_weight(weight):
    if 'kg' in weight:
        weight_value = re.findall(r"(\d+)", weight)[0]
    elif 'lbs' in weight:
        weight_value = float(re.findall(r"(\d+)", weight)[0]) * 0.453592  # Convert pounds to kilograms
    else:
        weight_value = weight
    return weight_value

data['Weight'] = data['Weight'].apply(convert_weight).astype(float)
```

```python
data['Weight']
```

```python
import re

def convert_height(height):
    if "'" in height:
        feet, inches = re.findall(r"(\d+)'(\d+)", height)[0]
        height_cm = int(feet) * 30.48 + int(inches) * 2.54
    else:
        height_cm = re.findall(r"(\d+)", height)[0]
    return int(height_cm)

data['Height'] = data['Height'].apply(convert_height).astype(float)
```
